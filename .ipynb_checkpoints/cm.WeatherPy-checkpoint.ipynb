### Three+ Observable Trends
1. Unsurprisingly, the data does show that a city's daily maximum temperature is strongly correlated with latitude. Solar energy has to travel through more atmosphere at the poles, and less of it reaches the surface of the earth. Because it is currently summer in the Northern Hemisphere, the max temperatures are clustered closer to the Tropic of Cancer than the equator.
1. "Cloudiness" is apparently most frequently reported at 0%, 20%, 40%, 75%, and 100%—creating unnatural bunchings of data at these values.
1. Humidity doesn't show much correlation with latitude mostly because the information is reported as *relative* humidity—the percentage of moisture that surface air can hold at a given temperature and pressure. Absolute humidity would absolutely (ha!) have a correlation with latitude given that warmer air can hold more mosture.
1. Despite selecting latitudes and longitutes at random, the distribution of nearest cities found by citypy favored the Northern Hemisphere. The x-axes of all the suggested plots show this, but I also produced a box plot. 68% of the Earth's land area is found in the Northern Hemisphere—as well as about 90% of its population—so this is unsurprising.

```python
# dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from datetime import datetime as dt
import openweathermapy as owm

# import API key
from api_keys import api_key

# incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# output File (CSV)
output_data_file = "cities.csv"

# range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to our cities list
    if city not in cities:
        cities.append(city)

# print the city count to confirm sufficient count
print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# empty list to hold the weather data
weather_data = []

# settings dictionary for openweathermapy
settings = {'units':'imperial','APPID':api_key}

# chunk the cities into sets of 20
num_sets = len(cities)//20

# retrieve the data from the OpenWeatherMap API via the openweathermapy wrapper
# begin feedback to the terminal
print('Beginning Data Retrieval\n'+'-'*30)

# loop through the sets
for sets in range(num_sets+1):    
    #loop through the 20 cities in each set
    for record in range(20):
        index = sets*20 + record
        # break from the loop when we get to the last city
        if index >= len(cities):
            break
        # provide feedback to the terminal while retrieving data
        print(f'Processing Record {record+1} of set {sets+1} | {index}:{cities[index]}')
        # handle the error that will occur if openweathermap doesn't have data for a citypy city
        try:
            data = owm.get_current(cities[index], **settings)
            weather_data.append(data)
        except:
            print('City not found. Skipping...')
# end feedback to the terminal           
print('-'*30+'\nData Retrieval Completed\n'+'-'*30)


# the API calls in this block take a few minutes to execute!
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# extract the desired columns of data
summary = ['name','clouds.all','sys.country','dt','main.humidity','coord.lat','coord.lon','main.temp_max','wind.speed']
weather_extract = [city_weather(*summary) for city_weather in weather_data]

# make the data frame
city_df = pd.DataFrame(weather_extract)

# name the columns
city_df.columns =['City', 'Cloudiness', 'Country', 'Date', 'Humidity', 'Latitude', 'Longitude', 'Max Temp', 'Wind Speed']

# save the data frame to csv
city_df.to_csv(output_data_file,index=False)

# city_df.count()
print(city_df.count())

# view data frame
city_df.head()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

```python
# get the analysis date from timestamp data
date = max(city_df['Date'])
year = dt.utcfromtimestamp(date).year
month = dt.utcfromtimestamp(date).month
day = dt.utcfromtimestamp(date).day
```

#### Latitude vs. Temperature Plot

```python
# draw a scatterplot
plt.scatter(city_df['Latitude'],city_df['Max Temp'],edgecolors='black',color='deepskyblue')

# add other plot features
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title(f'City Latitude vs. Max Temperature ({month}/{day}/{year})')

# save the plot
plt.savefig(f'Latitude v Temperature {month}-{day}-{year}.png',dpi=600)
```

#### Latitude vs. Humidity Plot

```python
# draw a scatterplot
plt.scatter(city_df['Latitude'],city_df['Humidity'],edgecolors='black',color='deepskyblue')

# add other plot features
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Relative Humidity (%)')
plt.title(f'City Latitude vs. Humidity ({month}/{day}/{year})')

# save the plot
plt.savefig(f'Latitude v Humidity {month}-{day}-{year}.png',dpi=600)
```

#### Latitude vs. Cloudiness Plot

```python
# draw a scatterplot
plt.scatter(city_df['Latitude'],city_df['Cloudiness'],edgecolors='black',color='deepskyblue')

# add other plot features
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.title(f'City Latitude vs. Cloudiness ({month}/{day}/{year})')

# save the plot
plt.savefig(f'Latitude v Cloudiness {month}-{day}-{year}.png',dpi=600)
```

#### Latitude vs. Wind Speed Plot

```python
# draw a scatterplot
plt.scatter(city_df['Latitude'],city_df['Wind Speed'],edgecolors='black',color='deepskyblue')

# add other plot features
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.title(f'City Latitude vs. Wind Speed ({month}/{day}/{year})')

# save the plot
plt.savefig(f'Latitude v Wind Speed {month}-{day}-{year}.png',dpi=600)
```

```python
# draw a box plot
plt.figure(figsize=(5,8))
plot1 = plt.boxplot(city_df['Latitude'],widths = .4,showfliers=True,patch_artist=True)

# add other plot features
plt.ylabel('Latitude')
plt.title('Distribution of City Latitudes')
plt.tick_params(bottom=False,labelbottom=False)
plt.grid()

for box in plot1['boxes']:
    box.set(facecolor = 'lightblue')

# save the plot
plt.savefig(f'Distribution of City Latitudes.png',dpi=600)
```
